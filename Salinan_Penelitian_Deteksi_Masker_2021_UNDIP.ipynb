#### Count Total Images in IMFD Directory  
Hitung jumlah gambar yang ada di folder hasil ektraksi file `IMFD.zip`.

```python
imfd_dir = base_dataframe + "/IMFD"
imfd_dir = pathlib.Path(imfd_dir)

imfd_count = len(list(imfd_dir.glob('*/*.jpg')))
print(imfd_count)
```

#### Count Each Class in IMFD  
IMFD terdiri dari 3 kelas berbeda, yaitu:  
1. Uncovered Chin (Mask_Nose_Mouth)  
2. Uncovered Nose (Mask_Mouth_Chin)  
3. Uncovered Mouth and Nose (Mask_Chin)  

Hitung jumlah masing-masing data pada setiap kelas.

```python
matching_patterns = [
                    '*/*Mask_Nose_Mouth.jpg',
                    '*/*Mask_Mouth_Chin.jpg',
                    '*/*Mask_Chin.jpg'
                  ]
```

```python
sum = 0
for pattern in matching_patterns:
  count = len(list(imfd_dir.glob(pattern)))
  sum += count
  print(count)

print("Total images in IMFD: ", sum)
```
